```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\Gaurav verma\Downloads\employee_salary_dataset.csv")
```

```python
print("Dataset Loaded Successfully!")
df.head()
```

```python
plt.figure(figsize=(10, 5))
# Group by department, calculate mean, and sort descending
dept_salary = df.groupby('Department')['Monthly_Salary'].mean().sort_values(ascending=False).reset_index()
sns.barplot(data=dept_salary, x='Department', y='Monthly_Salary')
plt.title('Bar Chart: Average Monthly Salary by Department')
plt.ylabel('Average Monthly Salary ($)')
plt.xlabel('Department')
plt.tight_layout()
plt.savefig('bar_chart.png')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Monthly_Salary'], bins=20, kde=True, color='purple')
plt.title('Histogram: Distribution of Monthly Salary')
plt.xlabel('Monthly Salary ($)')
plt.ylabel('Frequency (Number of Employees)')
plt.tight_layout()
plt.savefig('histogram.png')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='Education_Level', y='Monthly_Salary')
plt.title('Box Plot: Salary Distribution by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Monthly Salary ($)')
plt.tight_layout()
plt.savefig('boxplot.png')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
exp_salary = df.groupby('Experience_Years')['Monthly_Salary'].mean().reset_index()
sns.lineplot(data=exp_salary, x='Experience_Years', y='Monthly_Salary', marker='o')
plt.title('Line Chart: Average Salary Trend by Experience Years')
plt.xlabel('Years of Experience')
plt.ylabel('Average Monthly Salary ($)')
plt.tight_layout()
plt.savefig('lineplot.png')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(data=df, x='Experience_Years', y='Monthly_Salary', hue='Gender', s=100, alpha=0.7)
plt.title('Scatter Plot: Experience vs Monthly Salary')
plt.xlabel('Years of Experience')
plt.ylabel('Monthly Salary ($)')
plt.tight_layout()
plt.savefig('Scatterplot.png')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
numeric_cols = df[['Experience_Years', 'Age', 'Monthly_Salary']]
correlation_matrix = numeric_cols.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Heatmap: Correlation Matrix')
plt.tight_layout()
plt.savefig('Heatmap.png')
plt.show()
```

```python
dept_counts = df['Department'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(
    dept_counts,
    labels=dept_counts.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.cm.Set3.colors
)

plt.title('Pie Chart: Employee Distribution by Department', fontsize=14)
plt.axis('equal')
plt.tight_layout()
plt.savefig('pie_chart.png')
plt.show()
```
